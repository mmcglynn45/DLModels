```python


import sklearn
import numpy as np
import pandas

import sklearn
from sklearn import tree
from sklearn import ensemble

import skimage.data
import skimage.transform
```

```python
# Load Data
import math

width = 160
height = 120

fname = "G:/Output/interpolated.csv"
base = "G:/Output/"

lineLimit = 5000
lineCount = 0
trainingIndex = int(lineLimit/2)

# Import Labels
labels = []
imagePaths = []
lastSpeed = 0

with open(fname) as f:
    content = f.readlines()
    for line in content:
        if lineCount == 0:
            lineCount+=1
            continue;
        line = line.split(",")
        speedString = (line[8])
        speed = float(speedString)
        trackingAngleString = (line[6])
        trackingAngle = float(trackingAngleString)
        filePath = (line[5])
        camera = (line[4])
        speedDelta = speed - lastSpeed
        if abs(trackingAngle)>1:
            labels.append(trackingAngle)
            #print(speed)
            #print(trackingAngle)
            picturePath = base+filePath
            #print(picturePath)
            imagePaths.append(picturePath)
            lineCount += 1
            lastSpeed = speed
            if lineCount > lineLimit:
                break;
print("Parsing complete")
trainingIndex = int(lineCount/2)
print(lineCount)
```

```python
images = []
for i in range(len(imagePaths)):
    loadPath = imagePaths[i]
    try:
        image = skimage.data.imread(loadPath)
        images.append(image)
        images[-1] = skimage.transform.resize(images[-1], (height, width))
        #images.append(skimage.data.imread(loadPath))
    except:
        print(loadPath)
```

```python

images = np.asarray(images)
imagesFlat = images.reshape(len(images),width*height*3)
```

```python
labels_a = np.array(labels)
images_a = np.array(imagesFlat)
print("labels: ", labels_a.shape, "\nimages: ", imagesFlat.shape)
```

```python
RegressorClass = labels
clf = ensemble.RandomForestRegressor(n_jobs=-1)
for round in range(0,1):
    print("training start")
    clf = clf.fit(imagesFlat[:trainingIndex], labels[:trainingIndex])
    print("training end")
```

```python
#Calculate MSE for testing        
print(clf.score(imagesFlat[trainingIndex:], labels[trainingIndex:]))
```

```python
totalSquaredError = 0
for i in range(trainingIndex, len(images)):
    imageToPredict = imagesFlat[i]
    imageToPredict= imageToPredict.reshape(1, -1)
    result = clf.predict(imageToPredict)
    guess = result
    actual = labels[i]
    print("Guess: " + str(guess) + "   Actual: " + str(actual))
    squaredError = abs(actual-guess)*abs(actual-guess)
    totalSquaredError += squaredError
averageSquaredError = math.sqrt(totalSquaredError/(len(images)-trainingIndex))
print(averageSquaredError)
```
